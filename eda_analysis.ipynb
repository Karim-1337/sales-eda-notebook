```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

candidates = [
    Path("data/sample_sales.csv"),
    Path("sample_sales.csv"),
    Path("/content/data/sample_sales.csv"),
    Path("/content/sample_sales.csv"),
]
DATA_PATH = next((p for p in candidates if p.exists()), None)
if DATA_PATH is None:
    raise FileNotFoundError(
        "sample_sales.csv not found. Upload it in Colab or clone/open this project folder."
    )
print("CSV:", DATA_PATH.resolve())
```

## Load and inspect


```python
df = pd.read_csv(DATA_PATH, parse_dates=["date"])
df.head()
```

```python
df.info()
df.describe()
```

## Revenue by region


```python
by_region = (
    df.groupby("region", as_index=False)["revenue"]
    .sum()
    .sort_values("revenue", ascending=False)
)
by_region
```

## Plots


```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].bar(by_region["region"], by_region["revenue"], color="#2563eb")
axes[0].set_title("Total revenue by region")
axes[0].tick_params(axis="x", rotation=25)

daily = df.sort_values("date")
axes[1].plot(daily["date"], daily["revenue"], marker="o", color="#059669")
axes[1].set_title("Daily revenue (sample)")
axes[1].tick_params(axis="x", rotation=35)
fig.tight_layout()
plt.show()
```
